```python
q1=data['sale_price'].quantile(0.25)
q3=data['sale_price'].quantile(0.75)
iqr=q3-q1
iqr
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
upper_limit,lower_limit
```

```python
def limit(value):
    if value>upper_limit:
        return upper_limit
    if value<lower_limit:
        return upper_limit
    else:
        return value
```

```python
data['sale_price']=data['sale_price'].apply(limit)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Y=data['sale_price']
X=scaler.fit_transform(data.drop(columns=['sale_price']))
X=pd.DataFrame(data=X,columns=data.drop(columns=['sale_price']).columns)
X.head()
```

# Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data=X
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

```python
VIF[VIF==VIF.max()].index[0]
```

```python
def remover(data):
    vif=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
    if vif.max()>5:
        print(vif[vif==vif.max()].index[0],'has been removed')
        data=data.drop(columns=vif[vif==vif.max()].index[0])
        return data
    else:
        print("No multicollinearity present anymore")
        return data
        
```

```python
for i in range(3):
    vif_data=remover(vif_data)
vif_data.head()
```

```python
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data.columns)
```

# Train/Test set

```python
X=vif_data
Y=data['sale_price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
X_train.shape, X_test.shape, Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr.coef_
```

```python
predictions=lr.predict(X_test)
lr.score(X_test,Y_test)
```

# Residuals

```python
residuals=predictions-Y_test
residual_table=pd.DataFrame({'residuals':residuals,'predictions':predictions})
residual_table=residual_table.sort_values(by='predictions')
```

```python
z=[i for i in range(int(residual_table['predictions'].max()))]
k=[0 for i in range(int(residual_table['predictions'].max()))]
```

```python
plt.figure(dpi=130,figsize=(17,7))
plt.scatter(residual_table['predictions'],residual_table['residuals'],color='red')
plt.plot(z,k,color='green',label="regression line")
plt.ylim(-800000,800000)
plt.xlabel("fitted points(ordered by predictions)")
plt.ylabel("residuals")
plt.title("residualplot")
plt.legend()
plt.show()
```

```python
plt.hist(residual_table['residuals'],bins=200,color="red")
plt.xlabel("residuals")
plt.ylabel("frequency")
plt.title("distribution of residuals")
plt.show()
```
